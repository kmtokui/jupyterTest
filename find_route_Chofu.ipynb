```python
import pandas as pd
```

```python
# csvファイルの読み込み
NUM_FILES = 6
directory = 'csv/Chofu/result/'
filename = 'r_biccamera'
df1 = pd.read_csv(directory + filename + '.csv')
filename = 'r_kaisatu'
df2 = pd.read_csv(directory + filename + '.csv')
filename = 'r_minamiguchi'
df3 = pd.read_csv(directory + filename + '.csv')
filename = 'r_parco'
df4 = pd.read_csv(directory + filename + '.csv')
filename = 'r_seiyu'
df5 = pd.read_csv(directory + filename + '.csv')
filename = 'r_trie'
df6 = pd.read_csv(directory + filename + '.csv')

df = [df1, df2, df3, df4, df5, df6]
```

```python
# 滞在時間を計算
for i in range(NUM_FILES):
    df[i]['duration'] = df[i]['out'] - df[i]['in']
    print(len(df[i]['duration']))
    
```

```python
# BLEデータを管理するクラス
# 更新時間とIDを追加すると更新時間で並べ替えてくれる

class BleData:
    def __init__(self, address, index, time):
        self.address = address
        self.history = {time:index}
        self.sequence = []
        
    def update(self, address, index, time):
        
        # check address
        if self.address != address:
            return False
        
        # check index
        for t, i in self.history.items():
            if i == index:
                return False
        
        # add new position
        self.history[time] = index
        
        return True
    
    def sort(self):
#         self.history = sorted(self.history.items())
        for t, i in sorted(self.history.items()):
            self.sequence.append(i)
        
    def get_next(self, index):
        next_index = -1
        for i in range(len(self.sequence)-1):
            next_index = self.sequence[i]
            
        return next_index
        
        
```

```python
# 他の地点と共通するaddressを見つける
ble_list = []

for j in range(NUM_FILES):
    
    for k in range(len(df[j]['address'])):
        addr = df[j]['address'][k]
        time = df[j]['in'][k]
#         print(addr, time)
        
        is_new = True
        for i in range(len(ble_list)):
            if ble_list[i].address == addr:
                is_new = False
                
                # リストにあれば更新
                ble_list[i].update(addr, j, time)
#                 print('update data')
    
        # リストになければ追加
        if is_new:
            ble_list.append( BleData(addr, j, time) )
#             print('add new data')

print(len(ble_list))
                
        
```

```python
# 通過時刻順に並べ替える
for data in ble_list:
    data.sort()
```
